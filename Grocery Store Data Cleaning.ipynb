```python
import pandas as pd
import numpy as np
```

# Grocery Stores
Isolating only the California grocery stores.

```python
grocery_stores = pd.read_csv('./data/grocery_stores.csv')
grocery_stores.head()
```

```python
ca_only = grocery_stores[grocery_stores.State == 'CA'].copy()
ca_only.head()
```

```python
ca_only['geoid'] = ("0" + ca_only['CensusTract'].astype(str))
```

```python
ca_only.to_csv('./data/ca_only_grocery_stores.csv')
```

```python
li = ca_only.loc[:, ['geoid', 'LowIncomeTracts']].copy()
li.to_csv('./data/low_income_geoid.csv')
```
